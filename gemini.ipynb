```python
import os
from dotenv import load_dotenv
from google import genai
```

```python
load_dotenv()
```

```python
gemini_key = os.getenv("GEMINI_API_KEY")
```

```python
client = genai.Client(api_key=gemini_key)
```

```python
response = client.models.generate_content(
    model="gemini-3-flash-preview",
    contents="나는 커피를 좋아해!"
)
print(response.text)
```

```python
response = client.models.generate_content(
    model="gemini-3-flash-preview",
    contents="내가 뭘 좋아한다고 했지?"
)
print(response.text)
```

```python
response = client.models.generate_content(
    model="gemini-3-flash-preview",
    contents=[
        {"role": "user", "parts": [{"text": "나는 커피를 좋아해"}]},
        {"role": "model", "parts": [{"text": "좋아하시는 커피 스타일을 알려주시면 그에 맞는 커피 추천이나 재미있는 이야기도 해드릴 수 있어요! 유독 좋아하는 카페나 원두가 있는지도 궁금하네요. :)"}]},
        {"role": "user", "parts": [{"text": "내가 뭘 좋아한다고 했지?"}]},
    ]
)
print(response.text)
```

```python
prompt = """
다음 주제로 네이버 블로그 글을 작성해주세요.

제목: 커피 원두 종류별 특징과 맛 비교
키워드: 커피원두

조건:
- 소제목(##)을 3~4개로 나눠서 작성
- 친근한 말투로 작성
- 1500자 이상
- 마크다운 형식으로 작성
"""
```
