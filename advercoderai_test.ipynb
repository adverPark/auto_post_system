```python
data = {
    "campaign_name": "API 테스트",
    "type": "INFO",
    "titles": "커피 원두 종류별 특징과 맛 비교///커피원두",
    "model_choice": "GEMINI30FLASH",
    "start_date": "2026-03-05T10:00:00",
    "publish_status": "DNP",
    "publish_interval": 0,
    "tone": "FRD"
}
```

```python
response = requests.post(
    f"{BASE_URL}/campaigns/",
    headers=headers,
    json=data
)

result = response.json()
```

```python
campaign_id = result["id"]
total_posts = result["total_posts"]
print(f"캠페인 ID: {campaign_id}")
print(f"총 포스트 수: {total_posts}")
print(f"상태: {result['status']}")
```

```python
print("글 작성 대기 중...")

while True:
    response = requests.get(
        f"{BASE_URL}/campaigns/{campaign_id}/posts/?status=COM",
        headers=headers
    )
    completed = response.json()["count"]
    
    print(f"  완료: {completed}/{total_posts}")
    
    if completed >= total_posts:
        print("모든 글 작성 완료!")
        break
    
    time.sleep(30)
```

```python
response = requests.get(
    f"{BASE_URL}/campaigns/{campaign_id}/posts/",
    headers=headers
)

posts = response.json()["results"]

for post in posts:
    print(f"ID: {post['id']}")
    print(f"제목: {post['title']}")
    print(f"상태: {post['status']}")
    print()
```
