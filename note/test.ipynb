```python
import requests, urllib.parse

keyword = "바이브코딩"
url = f"https://ac.search.naver.com/nx/ac?q={urllib.parse.quote(keyword)}&con=1&frm=nv&ans=2&r_format=json&r_enc=UTF-8&r_unicode=0&t_koreng=1&q_enc=UTF-8&st=100"
data = requests.get(url).json()

for item in data["items"][0]:
    print(f"  - {item[0]}")
```
