```python
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.naver.com")
```

```python
search_box = driver.find_element(By.CSS_SELECTOR, "#query")
```

```python
search_box.send_keys("지원금")
```

```python
search_btn = driver.find_element(By.CSS_SELECTOR, ".btn_search")
```

```python
search_btn.click()
```

```python
driver.get("https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query=%EC%A7%80%EC%9B%90%EA%B8%88&ackey=8oke92k2")
```

```python
posts = driver.find_elements(By.CSS_SELECTOR, '[data-template-id="ugcItem"]')
```

```python
title = posts[0].find_element(By.CSS_SELECTOR, '.sds-comps-text-type-headline1').text
```

```python
link = posts[0].find_element(By.CSS_SELECTOR, 'a').text
```

```python
print(len(link))
```

```python
blog_name = posts[0].find_element(By.CSS_SELECTOR, '.sds-comps-text-ellipsis.sds-comps-text-ellipsis-1').text
```

```python
for post in posts:
    title = post.find_element(By.CSS_SELECTOR, '.sds-comps-text-type-headline1').text
    link = post.find_element(By.CSS_SELECTOR, 'a').get_attribute('href')
    blog_name = post.find_element(By.CSS_SELECTOR, '.sds-comps-text-ellipsis.sds-comps-text-ellipsis-1').text
    print(title)
    print(link)
    print(blog_name)
    print("======")
```
